## DBPedia Spotlight

```python
text = """
Во втором туре выборов президента Российской академии наук победил академик Александр Сергеев, сообщил результаты подсчета голосов председатель счетной комиссии Юрий Балега.
На выборах президента РАН собран кворум.
Протокол счетной комиссии утвержден общим собранием, сообщает ТАСС.

Напомним, что во второй тур выборов вышли академики Александр Сергеев и Роберт Нигматулин. В итоге за первого кандидата, по словам главы счетной комиссии, было подано 1045 голосов, за второго - 412 голосов.
"""
```

```python
r = requests.get('http://model.dbpedia-spotlight.org/ru/spot', 
                 params={'text': text, 'confidence': 0.1}, 
                 headers = {'accept': 'application/json'})
```

```python
json.loads(r.text)
```

```python
r = requests.get('http://model.dbpedia-spotlight.org/ru/annotate', 
                 params={'text': text}, headers = {'accept': 'application/json'})
```

```python
json.loads(r.text)
```
