## Пример Quantile-quantile plot

```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import scipy.stats as stats
```

```python
# задаем набор 100 случайных значений с нормальным распределением
# https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html
# loc - среднее распределения
# scale - стандартное отклонение
# size - количество точек

measurements = np.random.normal( loc = 20, scale = 5, size=500 )
```

```python
measurements[:5]
```

```python
plt.hist( measurements, bins = 20 )

plt.show()
```

```python
# используем метод probplot для построения QQ-plot
stats.probplot( measurements, dist="norm", plot=plt )
plt.show()
```
