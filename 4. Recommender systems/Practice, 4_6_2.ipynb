# fastText

```python
FASTTEXT_BIN = "/Users/tural/ExternalProjects/fastText/fasttext"
EN_MODEL_BIN = "/Users/tural/Datasets/wiki.simple/wiki.simple.bin"
N_FEATURES = 300

TMP_INPUT_PATH = "/tmp/data.txt"
TMP_OUTPUT_PATH = "/tmp/data.out.txt"
```

```python
from subprocess import Popen, PIPE

def get_fasttext_features_for_data(data):
    print("Storing text")
    with open(TMP_INPUT_PATH, "w") as f:
        f.write("\n".join(data))

    print("Processing text")
    cmd = [FASTTEXT_BIN, "print-sentence-vectors", EN_MODEL_BIN]
    p = Popen(cmd, stdin=open(TMP_INPUT_PATH, "r"), stdout=open(TMP_OUTPUT_PATH, "w"), stderr=PIPE)
    p.communicate()
            
    print("Extending articles")
    vectors = []
    with open(TMP_OUTPUT_PATH, "r") as f:
        for line in f:
            v = list(map(float, line.strip().rsplit(" ", N_FEATURES)[1:]))
            vectors.append(v)
    return vectors
```

```python
data = ["hello world", "good bye my darling"]
```

```python
vectors = get_fasttext_features_for_data(data)
```
